```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def fteste1_2a(N, E0, nequi, npassos):
    nmedidas = npassos - nequi

    n1 = int(np.floor((N - E0)/2))
    E = N - 2 * n1
    ED = E0 - E
    s = -np.ones(N) # particulas na caixa 1
    s[n1:N] = 1 # particulas na caixa 2

    EDmedio = 0
    Emedio = 0

    for t in range(1, npassos + 1):
        for act in range(1, N + 1):

            ip = np.random.randint(1, N + 1)
            ip -= 1

            dE = - 2 * s[ip]

            if dE < 0:
                s[ip] = -s[ip] # mover a particula de uma caixa para a outra
                E += dE
                ED -= dE
                
            else:

                if dE <= ED:
                    s[ip] = -s[ip] # mover a particula de uma caixa para a outra
                    E += dE
                    ED -= dE

        if t > nequi:
            EDmedio += ED
            Emedio += E 
    
    Emedio = Emedio / nmedidas
    EDmedio = EDmedio / nmedidas

    return Emedio, EDmedio
```

```python
def fteste1_2b(N, E0, nequi, npassos):
    nmedidas = npassos - nequi

    n = np.zeros(2)
    en = np.array([-1, 1])

    n[0] = np.floor((N - E0) / 2)
    n[1] = N - n[0]

    E = n[1] - n[0]

    ED = E0 - E

    EDmedio = 0
    Emedio = 0

    for t in range(1, npassos + 1):
        for act in range(1, 3):
            ip = np.random.randint(0, 2)

            dE = -2*en[ip]

            if dE < 0:
                if n[0] < N:
                    n[1] -= 1 # retirar da caixa 2
                    n[0] += 1 # colocar na caixa 1
                    E += dE
                    ED -= dE
            else:
                if dE <= ED:
                    if n[0] >= 1:
                        n[1] += 1 # colocar na caixa 2
                        n[0] -= 1 # retirar da caixa 1
                        E += dE
                        ED -= dE                
        if t > nequi:
            EDmedio += ED
            Emedio += E

    Emedio = Emedio / nmedidas
    EDmedio = EDmedio / nmedidas

    return Emedio, EDmedio
```

```python
N = 100

E0s = np.arange(-N + N/10, -N + 9*N/10 + 1, N/10)

Ta = np.zeros(len(E0s))
Ema = np.zeros(len(E0s))

Tb = np.zeros(len(E0s))
Emb = np.zeros(len(E0s))

for i, E0 in enumerate(E0s):
    nequi = 1000
    npassos = 10000

    Ema[i], EDmedio = fteste1_2a(N, E0, nequi, npassos)
    Ta[i] = 2/np.log((EDmedio + 2)/EDmedio)
    print(f"a) Calculo para E0 = {E0} com T = {Ta[i]}")

    nequi = 10000
    npassos = 100000

    Emb[i], EDmedio = fteste1_2b(N, E0, nequi, npassos)
    Tb[i] = 2/np.log((EDmedio + 2)/EDmedio)
    print(f"b) Calculo para E0 = {E0} com T = {Tb[i]}\n")
```

```python
Ta_t = np.arange(0.1, np.max(Ta), 0.01)
Tb_t = np.arange(0.1, np.max(Tb), 0.01)
Ema_t = -N * np.tanh(1/Ta_t)
Emb_t = -N + 2/(np.exp(2/Tb_t) - 1)
plt.figure(0)
plt.plot(Ta, Ema, 'rx', Ta_t, Ema_t, 'b')
plt.title("descerniveis")
plt.figure(1)
plt.plot(Tb, Emb, 'rx', Tb_t, Emb_t, 'b')
plt.title("indescerniveis")
plt.show()
```
