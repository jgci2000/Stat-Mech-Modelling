```python
import numpy as np
import matplotlib.pyplot as plt
```

# Exercício 1

```python
def randsample(p):
    p = p / np.sum(p)
    u = np.random.rand()
    
    s = p[0]
    i = 0
    
    while u > s:
        i += 1
        s += p[i]
    
    return i
```

```python
def pX(x_vals):
    return 1 / (2 * np.sqrt(x_vals))
    
```

```python
n_pnts = int(1E4)

# Criação dos rectangulos para o método aceitação rejeição

# x_min não pode ser 0
x_min = 0.00001
x_max = 1

n_rect = 30
dx_r = (x_max - x_min) / n_rect
xi_r = np.arange(x_min, x_max, dx_r)
xf_r = xi_r + dx_r

h_r = np.zeros(n_rect)
for i in range(n_rect):
    x_vals = np.arange(xi_r[i], xf_r[i], 0.001)
    h_r[i] = np.max(pX(x_vals))

# Plot dos retangulos com a função

for i in range(n_rect):
    plt.plot([xi_r[i], xi_r[i]], [0, h_r[i]], '-r')
    plt.plot([xi_r[i], xf_r[i]], [h_r[i], h_r[i]], '-r')
    plt.plot([xf_r[i], xf_r[i]], [h_r[i], 0], '-r')

plt.figure(1)
plt.plot(np.arange(x_min, x_max, 0.001), pX(np.arange(x_min, x_max, 0.001)), '-b')
plt.show()

# Método aceitação rejeição

x_accepted = list()
xp, yp = np.zeros(n_pnts), np.zeros(n_pnts)

n, n_a = 0, 0
i = 0

Ai = (xf_r - xi_r) * h_r
p = Ai / np.sum(Ai)

for i in range(n_pnts):
    i_r = randsample(p)
    
    xp[i] = np.random.rand() * (xf_r[i_r] - xi_r[i_r]) + xi_r[i_r]
    yp[i] = np.random.rand() * h_r[i_r]
    n += 1
    
    if yp[i] <= pX(xp[i]):
        x_accepted.append(xp[i])
        n_a += 1

x_accepted = np.array(x_accepted)

print("Fração de número rejeitados -> {:.4f}".format(1 - n_a / n))

# Cálculo do Histograma normalizado

x_max = np.max(x_accepted)
x_min = np.min(x_accepted)
n_bins = 30
dx = (x_max - x_min) / n_bins

values, bins = np.histogram(x_accepted, bins=np.linspace(x_min+dx/2, x_max+dx/2, n_bins))
values = values / (dx * np.sum(values))

# Plot do histograma com a densidade de probabilidade

plt.figure(2)
plt.plot(bins[0:-1], values, '*r')
plt.plot(np.arange(x_min, x_max, 0.0001), pX(np.arange(x_min, x_max, 0.0001)), '-b')

plt.show()

```

# Exercício 2

```python
def distrib_p(p_vals, T):
    return (1 / np.sqrt(2 * np.pi * T)) * np.exp(- p_vals**2 / (2 * T))
```

```python
def demon(E0, N, t_max, p):
    ED = E0
    E = 0
    
    delta = 1
               
    for t in range(t_max):
        for i in range(N):
            idx_p = np.random.randint(0, N)
            dp = delta * (2 * np.random.rand(3) - 1)
            
            delta_E = np.dot(dp, dp) / 2 + np.dot(p[idx_p, :], dp)
            
            if delta_E < 0 or ED >= delta_E:
                p [idx_p, :] += dp
                ED += - delta_E
                E += delta_E
    
    return p
```

```python
N = 1000
t_max = 1000

# Cálculo para E0 = 100

E0 = 100
# Como inicialmente toda a energia do sistema está no demon,
# temos que começar com partículas com momento 0.
p = np.zeros((N, 3))

p_100 = demon(E0, N, t_max, p)
px_100 = p_100[:, 0] 

# Histograma
p_max = np.max(px_100)
p_min = np.min(px_100)
n_bins = 30
dp = (p_max - p_min) / n_bins

values, bins_100 = np.histogram(px_100, bins=np.linspace(p_min+dp/2, p_max+dp/2, n_bins))
values_100 = values / (dp * np.sum(values))

# Cálculo para E0 = 50

E0 = 50
p = np.zeros((N, 3))

p_50 = demon(E0, N, t_max, p)
px_50 = p_50[:, 0] 

# Histograma
p_max = np.max(px_50)
p_min = np.min(px_50)
n_bins = 30
dp = (p_max - p_min) / n_bins

values, bins_50 = np.histogram(px_50, bins=np.linspace(p_min+dp/2, p_max+dp/2, n_bins))
values_50 = values / (dp * np.sum(values))

# Plot dos histogramas e das distribuições

p_vals = np.arange(-1, 1, 0.001)
T = (2 * 100 / (3 * N))
p_100 = distrib_p(p_vals, T)

T = (2 * 50 / (3 * N))
p_50 = distrib_p(p_vals, T)

plt.figure(1)
line_100 = plt.plot(bins_100[:-1], values_100, 'rx', label="E0 = 100")
plt.plot(p_vals, p_100, '-r')
plt.title("Distribuição dos momentos para E0 = 100")

plt.figure(2)
line_100 = plt.plot(bins_100[:-1], values_100, 'rx', label="E0 = 100")
line_50 = plt.plot(bins_50[:-1], values_50, 'b*', label="E0 = 50")
plt.title("Distribuição dos momentos para E0 = 100 e E0 = 50")
plt.legend()

plt.plot(p_vals, p_100, '-r')
plt.plot(p_vals, p_50, '-b')

plt.show()
```
